##### Q3
Cabin의 첫 자리는 객실 유형을 그 다음 숫자는 객실의 위치를 나타냅니다.

결측이 아닌 Cabin만을 뽑고, 객식 유형을 분리하여 Cabin_Type 변수를 만들고,

객실 위치를 분리하여 Cabin_No 변수를 만듭니다.

```python
# 내 답변
cabin_filled = df_titanic[df_titanic['Cabin'].notna()]
cabin_filled['Cabin_Type'] = cabin_filled['Cabin'].apply(
    lambda x: x[:1]
)
cabin_filled['Cabin_No'] = cabin_filled['Cabin'].apply(
    lambda x: x[1:]
)
cabin_filled
```

```python
# 강의 답변
## explode : Transform each element of a list-like to a row.

df_titanic.loc[df_titanic['Cabin'].notna(), 'Cabin'].str.split(' ')\
    .explode()\
    .apply(
        lambda x: pd.Series((x[:1], x[1:]), index=['Cabin_Type', 'Cabin_No'])
    )
    
```
